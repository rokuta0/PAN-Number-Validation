```python
df['PAN_Number'] = df['Pan_Numbers'].replace('NAN', pd.NA)
```

```python
df['PAN_Number']
```

```python
pan_pattern = r'^[A-Z]{5}[0-9]{4}[A-Z]$'
```

```python
def validating_pan(pan):
    if pd.isna(pan):
        return "Missing"

    if not re.match(pan_pattern,pan):
        return 'Invalid'

    for i in range(len(pan)-1):
        if pan[i]==pan[i+1]:
            return 'Invalid'

    if pan[:5]==''.join(chr(ord(pan[0])+i) for i in range(5)):
        return "Invalid"

    if pan[5:9] in ["0123", "1234", "2345", "3456", "4567", "5678", "6789"]:
        return "Invalid"

    return 'Valid'
```

```python
df['PAN_status']= df['PAN_Number'].apply(validating_pan)
```
